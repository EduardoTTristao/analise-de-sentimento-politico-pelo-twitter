```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
```

```python
dfs = {}
```

```python
for c in ['bolsonaro', 'lula', 'ciro', 'simone tebet', 'bivar', 'Luiz Felipe d‚ÄôAvila', 'janones', 'leonardo pericles', 'pablo ma√ßal', 'vera lucia', 'sofia manzano', 'eymael']:
    dfs[c] = pd.read_csv(f'dados/ultimos_100_tweets/{c}_tweets.csv')
    dfs[c] = dfs[c].dropna(subset=['text'])
```

```python
classificacoes = {}
```

```python
print(dfs.keys())
```

```python
c = list(dfs.keys())
```

```python
index_politico = int(input("Index do politico desejado: ")) 
classificacoes[c[index_politico]] = [] 
for tweet in dfs[c[index_politico]]['text']: 
    classificacoes[c[index_politico]].append(input(str(tweet)+" classifica√ß√£o: "))
```

```python
print(classificacoes[c[index_politico]])
print(len(classificacoes[c[index_politico]]))
```

```python
dfc = pd.DataFrame(classificacoes)
```

```python
#dfc.loc[181,'bivar'] = 'NT'
dfc = dfc.replace(['1','2','3'], ['NG','NT','PS'])
```
